```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV into a pandas DataFrame                                        
data = pd.read_csv('klimakongress_stats.csv')                             
                                                                          
# Convert the unix timestamp to a more readable format for matplotlib     
data['unixmilli'] = pd.to_datetime(data['unixmilli'], unit='ms')          
                                                                          
# Set the unixmilli column as index                                       
data.set_index('unixmilli', inplace=True)                                 
                                                                          
# Create the line plot                                                    
plt.figure(figsize=(12, 8)) # Set the figure size                         
                                                                          
# Plot each line                                                          
plt.plot(data['providers'], label='Providers')                            
plt.plot(data['workers'], label='Workers')                                
plt.plot(data['requests/sec'], label='Requests/sec')                      
                                                                          
# Add a legend                                                            
plt.legend()                                                              
                                                                          
# Label the axes                                                          
plt.xlabel('Time')                                                        
plt.ylabel('Count')                                                       
                                                                          
# Show the plot                                                           
plt.show()                                            
```
